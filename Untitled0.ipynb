```python
df['rate'].fillna(df['rate'].mode()[0], inplace=True)

df['location'].fillna('Unknown', inplace=True)

df['cuisines'].fillna('Unknown', inplace=True)

df['dish_liked'].fillna('Not Mentioned', inplace=True)

df['listed_in(type)'].fillna('Others', inplace=True)

df['approx_cost(for two people)'].fillna(
    df['approx_cost(for two people)'].mode()[0],
    inplace=True
)
```

```python
top_locations = df['location'].value_counts().head(10)
print(top_locations)
```

```python
import matplotlib.pyplot as plt

top_locations.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Restaurant Locations")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
top_cuisines = df['cuisines'].value_counts().head(10)
print(top_cuisines)
```

```python
top_cuisines.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
df[['rate','approx_cost(for two people)']].head()
```

```python
df['rate'] = df['rate'].str.replace('/5', '', regex=False)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
df['rate'].head()
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str)

df['approx_cost(for two people)'] = (
    df['approx_cost(for two people)']
    .str.replace(',', '')
)

df['approx_cost(for two people)'] = pd.to_numeric(
    df['approx_cost(for two people)'],
    errors='coerce'
)
```

```python
df[['rate','approx_cost(for two people)']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.scatterplot(
    x='approx_cost(for two people)',
    y='rate',
    data=df
)

plt.title("Price vs Rating")
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(df['cuisines'].dropna())

wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color='white'
).generate(text)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
